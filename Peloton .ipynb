# Drop vs Del

drop operates on both columns and rows; del operates on column only

drop can operate on multiple items at a time; del operates only on one at a time

drop can operate in-place or return a copy; del is an in-place operation only

```python
# Only delete one col, error if try more than one
del df['Avg. Heartrate']
```

```python
df.sample(1)
```

```python
# Drop using iloc 
# df.drop(df.iloc[:, 1:3], inplace = True, axis = 1) 
df.drop(df.iloc[:, 1:2], inplace = True, axis = 1)
```

```python
df.head(1)
```

```python
# drop using iloc and ix
# did not work, ix is deprecated

# df.drop(df.ix[:, 'Avg. Incline' : 'Avg. Pace (min/mi)'].columns, axis = 1)
```

```python
# Did not use inplace = True and in next cell the dropped columns are back
df.drop(['Avg. Incline','Avg. Pace (min/mi)'], axis = 1)
```

```python
df.head()
```

```python
# drop columns with inplace

df.drop(['Avg. Incline','Avg. Pace (min/mi)'], inplace = True, axis = 1)
```
